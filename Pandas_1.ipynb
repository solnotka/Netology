## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd
```

```python
power = pd.read_csv(r'c:\Netology_materials\power.csv')
```

```python
power.head(20)
```

```python
#Выясняю, как пишутся нужные страны
power[ power['country'].str.contains('L') ]['country'].unique()
```

```python
power[ power['country'].str.contains('Es') ]['country'].unique()
```

```python
#отфильтровываю по странам
filtered_power = power[ (power.country=='Lithuania') | (power.country=='Latvia') | (power.country =='Estonia') ]
filtered_power
```

```python
#Отфильтровываю по категориям
filtered_power = filtered_power[ (filtered_power.category==4) | (filtered_power.category==12) | (filtered_power.category==21) ]
```

```python
filtered_power.info()
```

```python
filtered_power['year'].unique()
```

```python
#Фильтрую по годам
filtered_power = filtered_power[ (filtered_power.year < 2011) &  (filtered_power.year > 2004) ]
filtered_power.sort_values('quantity')
```

```python
filtered_power['year'].unique()
```

```python
#Убираю отрицательные значения quantity
filtered_power = filtered_power[ filtered_power.quantity > 0 ]
filtered_power
```

```python
#Получаю сумму
filtered_power['quantity'].sum()
```
