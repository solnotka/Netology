## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

* если источник traffic_source равен yandex или google, то в source_type ставится organic
* для источников paid и email из России - ставим ad
* для источников paid и email не из России - ставим other
* все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log.info()
```

```python
def source_type(df):
    traffic_source = df['traffic_source']
    region = df['region']
    if traffic_source == 'yandex' or traffic_source == 'google':
        return 'organic'
    elif traffic_source == 'paid' and region == 'Russia' or traffic_source == 'email' and region == 'Russia':
        return 'ad'
    elif traffic_source == 'paid' and region != 'Russia' or traffic_source == 'email' and region != 'Russia':
        return 'other'
    else:
        return traffic_source
```

```python
log['source_type'] = log.apply(source_type, axis=1)
log.head(15)
```

```python
log.loc[log.traffic_source.str.contains('email|paid')].groupby(['region', 'traffic_source', 'source_type']).count()
```

```python
log.groupby(['traffic_source', 'source_type']).count()
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
ratings.head()
```

```python
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings.head()
```

```python
moviefans = (ratings.groupby('userId').agg({'rating': 'count', 'timestamp': lambda x: x.max() - x.min()}).
             rename(columns={'rating': 'ratings_count', 'timestamp': 'user_lifetime'}))
moviefans = moviefans[moviefans.ratings_count > 100]
moviefans
```

```python
moviefans['user_lifetime'].mean()
```
