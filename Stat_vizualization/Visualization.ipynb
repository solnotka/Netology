```python
columns = {'Country': 'Country or region', 'Happiness Rank': 'Overall rank', 'Happiness Score': 'Score',
            'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)': 'Healthy life expectancy',
                    'Freedom': 'Freedom to make life choices', 'Trust (Government Corruption)': 'Perceptions of corruption'}
```

```python
happiness_2015['Year'] = 2015
happiness_2016['Year'] = 2016
happiness_2017['Year'] = 2017
happiness_2018['Year'] = 2018
happiness_2019['Year'] = 2019
```

```python
happiness_2015.rename(columns=columns, inplace=True)
happiness_2015.info()
```

### Гистограмма. Распределение оценок счастья

```python
# Вариант с группировкой по странам и усредненными значениями
happiness_complete.groupby('Country or region').mean()['Score'].plot(kind='hist',
                                                                     title='Happiness distribution',
                                                                     xticks=range(1, 11))
```

В целом люди склонны оценивать свой уровень счастья как средний. Чаще всего встречаются оценки в диапазоне от 5 до 6 баллов. В таблице приводятся усредненные оценки счастья, видимо, поэтому нет оценок ниже тройки и выше восьмерки. 

```python
# Вариант без группировки по странам. Оценка за каждый год учитывается отдельно
happiness_complete['Score'].plot(kind='hist', title='Happiness distribution', xticks=range(1, 11))
```

На гистограмме с большим количеством значений две вершины - 4 и 6 баллов. Таким образом, по этому графику самые частые оценки - "немного выше среднего" и "немного ниже среднего".

```python
# Вариант по 2019 г.
happiness_2019['Score'].plot(kind='hist', title='Happiness distribution', xticks=range(1, 11))
```

Те же тенденции, что на предыдущем графике, только выражены более явно: большинство значений - "немного выше среднего" и "немного ниже среднего", между ними провал.

### График счастья в странах СНГ
Сравню уровень счастья в России, Украине и Беларуси за рассматриваемый период.

```python
RUB_happiness = happiness_complete[happiness_complete['Country or region'].isin(['Russia', 'Ukraine', 'Belarus'])]
RUB_happiness = RUB_happiness.pivot_table(index='Year', columns='Country or region', values='Score')
RUB_happiness.plot(title='Happiness in RUB')
```

Из графика видно, что самая несчастная страна - Украина, а самая счастливая - Россия. Пик этих состояний пришелся на 2017 год. После этой точки у Украины наметилась тенденция к росту, а у России и Беларуси идет спад. Белорусы в 2015 году были самыми счастливыми, но с тех пор все грустнеют и грустнеют.

### Boxplot
Посмотрю на значения столбца "GDP per capita" в России, Украине и Беларуси. Для сравнения добавлю Финляндию, которая находится в топе самых счастливых стран.

```python
RUBF_happiness = happiness_complete[happiness_complete['Country or region'].isin(['Russia', 'Ukraine', 'Belarus', 'Finland'])]
RUBF_GDP = RUBF_happiness.pivot_table(index='Year', columns='Country or region', values='GDP per capita')
sns.boxplot(data=RUBF_GDP)
```

Из графика видно, что ВВП - достаточно стабильный показатель. Выбросов нет, но учитывая, что для каждой страны анализируется по пять значений, это, наверное, естественно. Показатели ВВП, кажется, хорошо коррелируют с уровнем счастья: Украина опять в самом низу.

Сделаю такие же графики для уровня свободы и восприятия коррупции в этих странах.

```python
RUBF_freedom = RUBF_happiness.pivot_table(index='Year', columns='Country or region', values='Freedom to make life choices')
sns.boxplot(data=RUBF_freedom)
```

```python
RUBF_corruption = RUBF_happiness.pivot_table(index='Year', columns='Country or region', values='Perceptions of corruption')
sns.boxplot(data=RUBF_corruption)
```

На графике свободы впервые появляются выбросы: у Беларуси какая-то нестабильная ситуация с внутренней свободой. При этом общее положение стран не меняется.

После того, как я увидела график восприятия коррупции, мне становится понятно, что оказывается имелось в виду отсутствие коррупции и связь этого показателя со счастьем ожидается прямая. Россия оказалась по этому показателю в самом низу, примерно на одном уровне с Украиной. При этом оценки России по этому показателю похоже вообще не варьируют. И в целом по коррупции самые маленькие коробочки из всех трех графиков. Наверное, это значит, что данный показатель самый стабильный из трех рассмотренных.
